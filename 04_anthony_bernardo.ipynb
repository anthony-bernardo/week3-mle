## Exercice 3

```python
pas = 0.00001

def descent_update_x0(N, y_chapeau, y):
    normalizer = (1/N)
    return pas * normalizer * np.sum(y_chapeau - y)

def descent_update_x1(N, y_chapeau, y, x1):
    normalizer = (1 / N)
    return pas * normalizer * np.sum(y_chapeau - y) * x1
    
def h8(x1, theta):
    z = np.matrix([[1], [x1]])
    h2 = theta.T
    h3 = h2.dot(z)
    return np.asscalar(h3) # convert matrix 1,1 to scalar

import time

theta = np.matrix([[100], [100]])

thetas = []
errors = []
import random

# 100x (arbitraire)
for i in range(0, 100):
    index = random.randrange(0,200)
    
    A = descent_update_x0(X.shape[0], h8(x[index], theta), y[index])
    B = descent_update_x1(X.shape[0], h8(x[index], theta), y[index], x[index])

    print(A)
    print(B)
    
    theta[0] = theta[0] - A
    theta[1] = theta[1] - B
    
    print(theta[0])
    print(theta[1])
    
    plt.scatter(x, y)
    new_x = np.linspace(0, 250, 50) 
    plt.plot(new_x, h8(new_x, theta), color="red")
    plt.draw()
    plt.pause(0.0001)
    plt.clf()
    
    thetas.append(theta)
    
    error = cost_function(X.shape[0], h8(x[index], theta), y[index])
    errors.append(error)
    

    
```

```python
imin = np.argmin(errors)
print("-----")
print(f"epoch du min : {imin}")
print(f"error du min : {errors[imin]}")
print(f"thetas du min :")
print(thetas[imin])

plt.plot(errors,color="red")
```
